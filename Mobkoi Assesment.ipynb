   #### Click through rate = clicks/impressions

```python
merged_csv['ClickThroughRate'] = merged_csv['Clicks'] / merged_csv['Impressions'] * 100
merged_csv.head()
```

#### Video completion rate = video completions/Video starts

```python
merged_csv['VideoCompletionRate'] = merged_csv['VideoCompletions'] / merged_csv['VideoStarts'] * 100
merged_csv.head()
```

#### Video start rate = video starts/Impressions 

```python
merged_csv['VideoStartRate'] = merged_csv['VideoStarts'] / merged_csv['Impressions'] * 100
merged_csv.head()
```

```python
# Save new manipulated df and save as df
merged_csv.to_csv("FinalDf.csv", index=False)

df = pd.read_csv("/Users/Dev/Desktop/Mobkoi /CSV Files/FinalDf.csv")
```
